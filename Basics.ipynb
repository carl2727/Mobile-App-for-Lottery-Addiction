```python
import pandas as pd

lc = pd.read_csv("649.csv")

def factorial(n):
    final_product = 1
    for i in range(n, 0, -1):
        final_product *= i
    return final_product

def combinations(n, k):
    return int(factorial(n)/factorial(k)/factorial(n-k))

def one_ticket(l):
    p = combinations(49, 6)
    q = l/p
    return q

print("There are {} different possible outcomes for a lottery with 49 numbers. The chance of winning the lottery with one combination is {:.8f} %\n\n".format(combinations(49, 6), one_ticket(1)))

print(lc.shape, lc.head(3), lc.tail(3))

def extract_numbers(i):
    numbers = i[4:10]
    numbers_set = set(numbers.values)
    return numbers_set  
    

winning = lc.apply(extract_numbers, axis=1)
print(winning, "\n\n")


def check_historical_occurence(l, s):
    user_set = set(l)
    occur = user_set == s
    n_o = occur.sum()
    
    if n_o == 0:
        print("The lottery numbers {} were never drawn in history.".format(l))
    elif n_o == 1:
        print("The lottery numbers {} were already drawn one time in history.".format(l))
    else:
        print("The lottery numbers {} were already drawn {} time in history.".format(l, n_o))
    
l1 = [3, 7, 43, 47, 22, 26]    

check_historical_occurence(l1, winning)

def multi_ticket_probability(i):
    p_o = combinations(49, 6)
    c = i/p_o
    print("The chance of winning the lottery with {} tickets is {:.8f} %\n\n".format(i, c))

multi_ticket_probability(20)
```
